```python
# ============================================================
# STEP 1: Install required libraries
# - datasets: load and manage ML datasets (e.g. MRPC)
# - evaluate: compute metrics like accuracy and F1
# - transformers: provides BERT and tokenizer
# - accelerate: handles multi-GPU / multi-device training
# ============================================================
!pip install datasets evaluate transformers[sentencepiece]
!pip install accelerate
# To run the training on TPU, you will need to uncomment the following line:
# !pip install cloud-tpu-client==0.10 torch==1.9.0 https://storage.googleapis.com/tpu-pytorch/wheels/torch_xla-1.9-cp37-cp37m-linux_x86_64.whl
```

```python
# ============================================================
# STEP 2: Load dataset and tokenize
# MRPC = Microsoft Research Paraphrase Corpus
# Task: predict if two sentences have the same meaning (1) or not (0)
# checkpoint: the pretrained BERT model we will fine-tune
# tokenize_function: converts raw text into token IDs BERT understands
# DataCollatorWithPadding: pads sequences in a batch to equal length
# ============================================================
from datasets import load_dataset
from transformers import AutoTokenizer, DataCollatorWithPadding

raw_datasets = load_dataset("glue", "mrpc")
checkpoint = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)


def tokenize_function(example):
    return tokenizer(example["sentence1"], example["sentence2"], truncation=True)


tokenized_datasets = raw_datasets.map(tokenize_function, batched=True)
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
# ============================================================
# STEP 3: Clean up the dataset columns
# Remove raw text columns (model does not need them anymore)
# Rename 'label' to 'labels' (PyTorch expects this name)
# set_format('torch'): returns PyTorch tensors instead of lists
# ============================================================
tokenized_datasets = tokenized_datasets.remove_columns(["sentence1", "sentence2", "idx"])
tokenized_datasets = tokenized_datasets.rename_column("label", "labels")
tokenized_datasets.set_format("torch")
tokenized_datasets["train"].column_names
```

```python
# Expected output — the four columns the model will receive:
# attention_mask, input_ids, labels, token_type_ids
["attention_mask", "input_ids", "labels", "token_type_ids"]
```

```python
# ============================================================
# STEP 4: Create DataLoaders (batch iterators)
# DataLoader groups examples into batches of 8
# shuffle=True on training set prevents the model memorising order
# collate_fn=data_collator handles padding within each batch
# ============================================================
from torch.utils.data import DataLoader

train_dataloader = DataLoader(
    tokenized_datasets["train"], shuffle=True, batch_size=8, collate_fn=data_collator
)
eval_dataloader = DataLoader(
    tokenized_datasets["validation"], batch_size=8, collate_fn=data_collator
)
```

```python
# Quick sanity check — grab one batch and print its tensor shapes
# e.g. input_ids shape should be [8, seq_len]
for batch in train_dataloader:
    break
{k: v.shape for k, v in batch.items()}
```

```python
# ============================================================
# STEP 5: Load the pretrained BERT model
# from_pretrained: downloads BERT weights trained on Wikipedia + BookCorpus
# AutoModelForSequenceClassification adds a small classification
# layer on top of BERT automatically
# num_labels=2: two output classes (same meaning / different meaning)
# ============================================================
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(checkpoint, num_labels=2)
```

```python
# Quick forward pass test with one batch
# outputs.loss: how wrong the model is right now (before any training)
# outputs.logits.shape: should be [8, 2] — one score per class per example
outputs = model(**batch)
print(outputs.loss, outputs.logits.shape)
```

```python
# ============================================================
# STEP 6: Set up the optimizer
# AdamW is a popular optimizer for transformer fine-tuning
# lr=5e-5 (0.00005) is a standard starting learning rate for BERT
# ============================================================
from torch.optim import AdamW

optimizer = AdamW(model.parameters(), lr=5e-5)
```

```python
# ============================================================
# STEP 7: Set up the learning rate scheduler
# 'linear' schedule: starts at lr=5e-5 and gradually reduces to 0
# num_warmup_steps=0: no warmup (can add warmup for more stability)
# num_training_steps = epochs * batches_per_epoch
# ============================================================
from transformers import get_scheduler

num_epochs = 3
num_training_steps = num_epochs * len(train_dataloader)
lr_scheduler = get_scheduler(
    "linear",
    optimizer=optimizer,
    num_warmup_steps=0,
    num_training_steps=num_training_steps,
)
print(num_training_steps)
```

```python
# ============================================================
# STEP 8: Move model to GPU if available, otherwise CPU
# Training on GPU is ~10-50x faster than CPU
# ============================================================
import torch

device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
model.to(device)
device
```

```python
# ============================================================
# STEP 9: Training loop
# For each epoch (full pass over the data):
#   1. Feed a batch into the model (forward pass)
#   2. Compute the loss (how wrong the prediction was)
#   3. loss.backward(): calculate gradients (which direction to adjust weights)
#   4. optimizer.step(): update the model weights
#   5. lr_scheduler.step(): reduce learning rate
#   6. optimizer.zero_grad(): clear gradients for next batch
# tqdm shows a live progress bar
# ============================================================
from tqdm.auto import tqdm

progress_bar = tqdm(range(num_training_steps))

model.train()
for epoch in range(num_epochs):
    for batch in train_dataloader:
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        loss = outputs.loss
        loss.backward()

        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()
        progress_bar.update(1)
```

```python
# ============================================================
# STEP 10: Evaluate the model on the validation set
# model.eval(): switches off dropout (used only during training)
# torch.no_grad(): disables gradient tracking to save memory
# argmax: pick the class with the highest score as the prediction
# metric.compute(): returns accuracy and F1 score
# ============================================================
import evaluate

metric = evaluate.load("glue", "mrpc")
model.eval()
for batch in eval_dataloader:
    batch = {k: v.to(device) for k, v in batch.items()}
    with torch.no_grad():
        outputs = model(**batch)

    logits = outputs.logits
    predictions = torch.argmax(logits, dim=-1)
    metric.add_batch(predictions=predictions, references=batch["labels"])

metric.compute()
```

```python
# ============================================================
# STEP 11: Clean re-run of training with slightly lower lr (3e-5)
# Resets the model and optimizer from scratch
# Useful for experimenting with different hyperparameters
# ============================================================
from transformers import AutoModelForSequenceClassification, get_scheduler
from torch.optim import AdamW

model = AutoModelForSequenceClassification.from_pretrained(checkpoint, num_labels=2)
optimizer = AdamW(model.parameters(), lr=3e-5)

device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
model.to(device)

num_epochs = 3
num_training_steps = num_epochs * len(train_dataloader)
lr_scheduler = get_scheduler(
    "linear",
    optimizer=optimizer,
    num_warmup_steps=0,
    num_training_steps=num_training_steps,
)

progress_bar = tqdm(range(num_training_steps))

model.train()
for epoch in range(num_epochs):
    for batch in train_dataloader:
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        loss = outputs.loss
        loss.backward()

        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()
        progress_bar.update(1)
```

```python
# ============================================================
# STEP 12: Wrap training in a function for Accelerate
# Accelerate handles multi-GPU / TPU without changing the training code
# accelerator.prepare(): wraps model, optimizer, and dataloaders
# accelerator.backward(loss): replaces loss.backward() for compatibility
# This function will be launched by notebook_launcher below
# ============================================================
def training_function():
    from accelerate import Accelerator
    from transformers import AutoModelForSequenceClassification, get_scheduler
    from torch.optim import AdamW
    from tqdm.auto import tqdm

    accelerator = Accelerator()

    model = AutoModelForSequenceClassification.from_pretrained(checkpoint, num_labels=2)
    optimizer = AdamW(model.parameters(), lr=3e-5)

    train_dl, eval_dl, model, optimizer = accelerator.prepare(
        train_dataloader, eval_dataloader, model, optimizer
    )

    num_epochs = 3
    num_training_steps = num_epochs * len(train_dl)
    lr_scheduler = get_scheduler(
        "linear",
        optimizer=optimizer,
        num_warmup_steps=0,
        num_training_steps=num_training_steps,
    )

    progress_bar = tqdm(range(num_training_steps))

    model.train()
    for epoch in range(num_epochs):
        for batch in train_dl:
            outputs = model(**batch)
            loss = outputs.loss
            accelerator.backward(loss)

            optimizer.step()
            lr_scheduler.step()
            optimizer.zero_grad()
            progress_bar.update(1)
```

```python
# ============================================================
# STEP 13: Launch training with Accelerate
# notebook_launcher runs training_function across available devices
# Works on CPU, single GPU, or multiple GPUs with no code changes
# ============================================================
from accelerate import notebook_launcher

notebook_launcher(training_function)
```
